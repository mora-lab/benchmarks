# Load pathway database
 
   Load the two pathways (MISMATCH REPAIR and REGULATION OF AUTOPHAGY) by "KEGG"(supplied with the package)

```r
read_gmt = function(file){
  if(!grepl("\\.gmt$",file)[1]){stop("Pathway information must be a .gmt file")}; 
  geneSetDB = readLines(file);                                    ##read in the gmt file as a vector of lines
  geneSetDB = strsplit(geneSetDB,"\t");                           ##convert from vector of strings to a list
  names(geneSetDB) = sapply(geneSetDB,"[",1);                     ##move the names column as the names of the list
  geneSetDB = lapply(geneSetDB, "[",-1:-2);                       ##remove name and description columns
  geneSetDB = lapply(geneSetDB, function(x){x[which(x!="")]});    ##remove empty strings
  return(geneSetDB);
}

pathwaylist = read_gmt("E:/SSbenchmark_test/data/c2.cp.kegg.v7.0.symbols.gmt");
PATHWAYS = list();
PATHWAYS$gs = pathwaylist;
PATHWAYS$pathwaynames = as.list(names(pathwaylist));
names(PATHWAYS$pathwaynames) = names(pathwaylist);
```

# Quantify deregulation of pathways

 Calculate the deregulation score by running pathifier

```r
result_GSE10245 = quantify_pathways_deregulation(GSE10245_input$data,
                                                 GSE10245_input$allgenes,
                                                 PATHWAYS$gs,
                                                 PATHWAYS$pathwaynames,
                                                 GSE10245_input$normals,
                                                 # maximize_stability = T,
                                                 attempts = 100,
                                                 min_std = GSE10245_input$min_std,
                                                 min_exp = GSE10245_input$min_exp)

### result.PDS = t(mapply(FUN = c, result$scores));
### colnames(result.PDS) = colnames(gsedata$data)
```

The deregulation scores are now in `result_GSE10245$scores`, ready for further analysis.
